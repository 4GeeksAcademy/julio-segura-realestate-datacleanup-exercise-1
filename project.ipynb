#### What can you say about the price per square meter between the towns of Getafe and Alcorcón?

```python
ciudades = ['Getafe', 'Alcorcón']
filtro_ciudades = dfresto[dfresto['level5'].isin(ciudades)]

filtro_getafe = filtro_ciudades[filtro_ciudades['level5'] == 'Getafe']
filtro_alcorcon = filtro_ciudades[filtro_ciudades['level5'] == 'Alcorcón']

precio_getafe = filtro_getafe['price']
precio_alcorcon = filtro_alcorcon['price']

superficie_getafe = filtro_getafe['surface']
superficie_alcorcon = filtro_alcorcon['surface']

print("La suma de los precios en Getafe es: ", precio_getafe.sum())
print("La suma de los precios en Alcorcón es: ", precio_alcorcon.sum())

print("La suma de la superficie en Getafe es: ", superficie_getafe.sum())
print("La suma de la superficie en Alcorcón es: ", superficie_alcorcon.sum())

precio_m2_getafe = precio_getafe.sum() / superficie_getafe.sum()
precio_m2_alcorcon = precio_alcorcon.sum() / superficie_alcorcon.sum()

print("El precio por m2 en Getafe es de: ", round(precio_m2_getafe, 2))
print("El precio por m2 en Alcorcóon es de: ", round(precio_m2_alcorcon, 2))
```

#### Reload the data, is the average of "Valdemorillo" and "Galapagar" prices the same?

```python
ciudades2 = ['Valdemorillo', 'Galapagar']
filtro_ciudades2 = ds[ds['level5'].isin(ciudades2)]

filtro_valdemorillo = filtro_ciudades2[filtro_ciudades2['level5'] == 'Valdemorillo']
filtro_galapagar = filtro_ciudades2[filtro_ciudades2['level5'] == 'Galapagar']

precio_valdemorillo = filtro_valdemorillo['price']
precio_galapagar = filtro_galapagar['price']

print("El precio medio en Valdemorillo es: ", round(precio_valdemorillo.mean(), 2))
print("El precio medio en Galapagar es: ", round(precio_galapagar.mean(), 2))
print("El precio medio no es el mismo en Valdemorillo y Galapagar. El precio medio es más caro en Valdemorillo")
```

```python
# Solucion chatgpt:

# Calculate the average price for each group
grouped = ds.groupby('level5')
average_prices = grouped['price'].mean()

# Compare the average prices of the two groups
group_a_avg_price = average_prices.loc['Valdemorillo']
group_b_avg_price = average_prices.loc['Galapagar']

print("Average price for Valdemorillo:", round(group_a_avg_price,2))
print("Average price for Galapagar:", round(group_b_avg_price,2))

if group_a_avg_price > group_b_avg_price:
    print("Valdemorillo has a higher average price.")
elif group_a_avg_price < group_b_avg_price:
    print("Galapagar has a higher average price.")
else:
    print("The average prices of Valdemorillo and Galapagar are equal.")
```
